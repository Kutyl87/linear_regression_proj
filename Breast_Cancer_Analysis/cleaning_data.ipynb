```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
breast_data = pd.read_csv("data.csv")
```

```python
breast_data.head()
```

```python
breast_data.isnull().sum()
```

```python
breast_data.shape
breast_data = breast_data.drop(["Unnamed: 32"], axis=1)
```

```python
breast_data.dtypes
```

```python
breast_data["diagnosis"] = breast_data["diagnosis"].apply(lambda x:1 if x=="M" else 0)
```

```python
# wyznaczenie granic interkwartylowych (IQR) dla każdej kolumny
Q1 = breast_data.quantile(0.25)
Q3 = breast_data.quantile(0.75)
IQR = Q3 - Q1

# wykrycie wartości odstających dla każdej kolumny
outliers = (breast_data.lt(Q1 - 2 * IQR)) | (breast_data.gt(Q3 + 2 * IQR))
outliers = outliers.drop(["id"],axis=1)
# outliers = outliers.drop(["id"])
# usunięcie wartości odstających z danych
clean_data = breast_data[~outliers.any(axis=1)]
clean_data.shape
```
